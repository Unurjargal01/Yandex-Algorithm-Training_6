## Task 1

### Problem: Ancestor Tree, distance to each node
![title](Problems/4_A.png)

### Solution:

```python
# Task 1
from collections import defaultdict
def traverse(tree, root):
    ans = [(root, 0)]
    h = 1
    q = set([root])
    while q:
        new_sons = set()
        
        for f in q: # fathers
            for s in tree[f]: # sons
                ans.append((s, h))
                new_sons.add(s)
        h+= 1
        q = new_sons
    return sorted(ans)
                
PATH = 'HW 4/'
TASK = 'A_'
test_data = ['inp1.txt', 'inp2.txt', 'inp3.txt']
for i in test_data:
    data = open(PATH+TASK+i, 'r')
    n = int(data.readline())
    tree = defaultdict(set)
    sons = set()
    for _ in range(n-1):
        son, father = data.readline().split()
        tree[father].add(son)
        sons.add(son)
    ancestor = (set(tree) - sons).pop()
    for i in traverse(tree, ancestor):
        print(*i)

    print ("\n\n")
```

## Task 3

### Problems: Least Common Ancestor (LCA)
![title](Problems/4_C.png)

### Solution:

```python
from collections import defaultdict
import sys
sys.setrecursionlimit(100000)

def traverse(tree, root):
    cur = set([root])
    if not tree[root]: 
        ans[root] = cur
        return set([root])
    for son in tree[root]:
        cur |= traverse(tree, son)
    ans[root] = cur
    return cur

def LCA(tree, left, right, ancestor):
    elems = set([left, right])
    cur = ancestor
    while elems <= ans[cur]:
        for son in tree[cur]:
            if elems <= ans[son]:
                cur = son
                break
        else:
            return cur
data = open("HW 4/C_inp1.txt", 'r')
n = int(data.readline())
tree = defaultdict(set)
sons = set()
for _ in range(n-1):
	son, father = data.readline().split()
	tree[father].add(son)
	sons.add(son)
ancestor = (set(tree) - sons).pop()
ans = defaultdict(set)
traverse(tree, ancestor)
for line in data.readlines():
    left, right = line.split()
    print(LCA(tree, left, right, ancestor))
```

## Task 4

### Problem: Create Binary Tree given operations ADD, SEARCH, PRINTTREE.
### When printing the tree, depth of node is defined by "."
![title](Problems/4_D.png)

### Solution:

```python
class Node:
    def __init__(self, n):
        self.val = n
        self.left = None
        self.right = None
    def __repr__(self):
        return f"Node({self.val})"

def search(node: Node, x: int):
    """
        return Node with value x or None
    """
    cur = node
    while cur.val is not None:
        if cur.val == x:
            break
        elif cur.val < x:
            if cur.right is None: cur.right = Node(None)
            cur = cur.right
        else:
            if cur.left is None: cur.left = Node(None)
            cur = cur.left
    return cur
def printTree(node, depth = ""):
    if node is None: return
    if node.val is None: return
    printTree(node.left, depth+'.')
    print(depth+str(node.val))
    printTree(node.right, depth+'.')
    
root = Node(None)
# data = open("HW 4/D_inp1.txt", 'r')
while True:
    try:
        cmd = input().split()
        # cmd = data.readline().split()
        if len(cmd) == 2:
            ans = search(root, int(cmd[1]))
            if ans.val == int(cmd[1]):
                print("ALREADY" if cmd[0] == "ADD" else "YES")
            else:
                if cmd[0] == 'ADD':
                    ans.val = int(cmd[1])
                print("DONE" if cmd[0] == "ADD" else "NO")
        elif len(cmd) == 1:
            printTree(root)
        else:
            break
    except:
        break
```
