## Task 1:

### Problem: Correct bracket?

![title](Problems/3_A.png)

### Solution:

```python
# Task 1:
def solve(s):
    stack = []
    opening = ['[', '{', '(']
    closing = {']': '[', '}': '{', ')': '('}
    
    for i in s:
        if i in opening:        
            stack.append(i)
        elif i in closing:
            if len(stack) == 0:
                return 'no'
            if closing[i] == stack[-1]:
                stack.pop()
            else:
                return 'no'
        else:
            return 'no'
    return 'no' if len(stack) != 0 else 'yes'
s = input()
print(solve(s))
```

## Task 2:

### Problem
![title](Problems/3_B.png)

### Solution:

```python
# Task 2:
n = int(input())
nums = list(map(int, input().split()))
ans = [-1] * n
# value, index
stack = []
for i in range(n):
    while stack:
        if stack[-1][0] <= nums[i]:
            break
        _, idx = stack.pop()
        ans[idx] = i
    stack.append((nums[i], i))
print(*ans)
```

### Solution:

```python
# Task 3
from collections import deque
n, k = map(int, input().split())
nums = list(map(int, input().split()))

queue = deque([nums[0]])
if k == 1:
	print(nums[0])
for i in range(1, n):
    if i >= k and nums[i-k] == queue[0]:
        queue.popleft()
    if queue:    
        if queue[-1] > nums[i]:
            while queue:
                if queue[-1] <= nums[i]:
                    break
                queue.pop()

    queue.append(nums[i])
    if i >= k-1:
        print(queue[0])
```

## Task 4:

### Problem: Postfix writing
![title](Problems/3_D.png)

### Solution:

```python
# Task 4:
s = input().split()
op = ['-', '+', '*', '/']
stack = []
for i in s:
    if i in op:
        b, a = stack.pop(), stack.pop()
        # print(a, b, i)
        stack.append(str(eval(a+i+b)))
    else:
        stack.append(i)
    # print(stack)
print(*stack)
```

### Problem: Given Lexicographic order on brackets, generate with length of n as minimum as possible
![title](Problems/3_F.png)

### Solution:

```python
# Task 6:
n = int(input())
p = {j: i for i, j in enumerate(input())}
s = input()
ans = s
stack = []
opp = {'[': ']', '(': ')'}
for i in s:
    if stack:
        if (stack[-1] == '[' and i == ']') or (stack[-1] == '(' and i == ')'):
            stack.pop()
            continue
    stack.append(i)
cnt = (n - len(s) - len(stack))
first, second = ('[', '(') if p['['] < p['('] else ('(', '[')
while cnt > 0:
    if stack:
        # closing the bracket is worth it
        closing = opp[stack[-1]]
        if p[closing] < p[first]:
            stack.pop()
            ans+=closing
            continue
    # if we still can open the bracket
    stack.append(first)
    ans+=first
    cnt -= 2
# Cleaning
while stack:
    ans += opp[stack.pop()]
print(ans)
```

## Task 7:

### Problems: Waiting minutes on the Casher Queue
![title](Problems/3_G.png)

### Solution:

```python
n, b = map(int, input().split())
nums = {i: j for i, j in enumerate(map(int, input().split()))}
queue = [n-1-i for i in range(n)]
ans = sum(nums.values())
for i in range(n):
    cnt = b
    u = queue.pop()
    while u <= i and cnt > 0:
        # Extra Minutes
        if nums[u] > cnt:
            nums[u] -= cnt
            ans += (i - u) * cnt
            cnt = 0
        elif nums[u] <= cnt:
            cnt -= nums[u]
            ans += (i - u) * nums[u]
            nums[u] = 0
            if not queue:
                break
            u = queue.pop()
    queue.append(u)
    # print(queue, nums)
ans += sum([nums[i] * (n-i) for i in queue])
print(ans)
```

```python
from collections import deque

n, H = map(int, input().split())
h = list(map(int, input().split()))
w = list(map(int, input().split()))

heights = dict()
# Unifying same heights
for i in range(n):
    heights[h[i]] = heights.get(h[i], 0) + w[i]

items = list(heights.items())
items.sort(key = lambda x: x[0])
min_max = float('inf') if items[0][1] < H else 0
cnt = items[0][1] # width of first element
start = 0
queue_diff = deque([0]) # max

for last in range(1, len(items)):
    cnt += items[last][1] # adding the width
    rm1 = items[last][0] - items[last-1][0] # Added Difference

    # Adding difference to queue_diff: keeping decreasing seq
    while queue_diff:
        if queue_diff[-1] >= rm1:
            break
        queue_diff.pop()
    queue_diff.append(rm1)        
	# Sliding Window
    try:
        while cnt >= H and start <= last:
            cnt -= items[start][1] # Reducing 
            start += 1 # Moving
            
            if queue_diff:
                min_max = min(min_max, queue_diff[0]) # If satisfies condition keeping the max.
                rm = items[start][0] - items[start-1][0] # removed height after sliding window shifted
                if rm == queue_diff[0]:
                    queue_diff.popleft() # What if queue_diff = [] only if one cnt == H 
            else:
                min_max = 0
                # queue_diff means cnt == H or cnt > H
    except:
        continue
print(min_max)
```
